```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit import Aer, execute
from qiskit.tools.visualization import plot_histogram
```

```python
#cutedge checker
def ccheck(qc, a, b, c): 
    qc.cx(a, c) 
    qc.cx(b, c)
    
#inverse checker:
def iccheck(qc, a, b, c): 
    qc.cx(b, c)
    qc.cx(a, c) 
```

```python
def diffusion(qc, q):
    qc.h(q[0:4])
    qc.barrier()
    qc.x(q[0:4])
    qc.barrier()
    qc.h(q[3])
    qc.mct([q[0],q[1],q[2]], q[3] , [q[7]], mode='basic')
    qc.h(q[3])
    qc.barrier()
    qc.x(q[0:4])
    qc.barrier()
    qc.h(q[0:4])
    qc.barrier()
```

```python
q = QuantumRegister(11)
c = ClassicalRegister(4)
qc = QuantumCircuit(q,c)
itera = 2 #number of iteration

#initialization stage:
qc.h(q[0:4])
qc.x(q[8])
qc.h(q[8])
qc.barrier()

for i in range(itera):
    ccheck(qc,q[0],q[1],q[4])
    ccheck(qc,q[0],q[2],q[5])
    ccheck(qc,q[0],q[3],q[6]) 
    qc.mct([q[4],q[5],q[6]], q[8] , [q[7]], mode='basic') 
    iccheck(qc,q[0],q[3],q[6])
    iccheck(qc,q[0],q[2],q[5])
    iccheck(qc,q[0],q[1],q[4])
    diffusion(qc, q)
qc.measure(q[0:4], c[0:4])
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=10000)
result = job.result()
count =result.get_counts()
print(count)
qc.draw(output='mpl')
```

```python
#Decompose the circuit by using the Unroller
from qiskit.transpiler import PassManager
from qiskit.transpiler.passes import Unroller
pass_ = Unroller(['u3', 'cx'])
pm = PassManager(pass_)
new_circuit = pm.run(qc) 
new_circuit.draw(output='mpl')
```
